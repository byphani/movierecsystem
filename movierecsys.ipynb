```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
credits = pd.read_csv("credits.csv")
```

```python
df1 =movies.merge(credits, on="title")
```

```python
df1 = df1[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
df1
```

```python
import ast
```

```python
def helper(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
df2 = df1.dropna(inplace = True)
```

```python
df1.dropna(inplace = True)
```

```python
df2 = df1
```

```python
df1['genres'] = df1['genres'].apply(helper)
```

```python
df1['keywords'] = df1['keywords'].apply(helper)
```

```python
df1
```

```python
def helper1(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
df1['cast'] = df1['cast'].apply(helper1)
```

```python
df1
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l        
```

```python
df1['crew'] = df1['crew'].apply(fetch_director)
```

```python
def rem_spaces(L):
    l = [];
    for i in L:
        l.append(i.replace(" ", ""))
    return l
```

```python
df1['cast'] = df1['cast'].apply(rem_spaces)
df1['crew'] = df1['crew'].apply(rem_spaces)
df1['genres'] = df1['genres'].apply(rem_spaces)
df1['keywords'] = df1['keywords'].apply(rem_spaces)
```

```python
df1
```

```python
df1['overview'] = df1['overview'].apply(lambda x:x.split())
```

```python
df1['tags'] = df1['overview'] + df1['genres'] + df1['keywords'] + df1['cast'] + df1['crew']
```

```python
df1
```

```python
df2 = df1.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
df2
```

```python
df2['tags'] = df2['tags'].apply(lambda x: " ".join(x))
```

```python
df2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vec_words = cv.fit_transform(df2['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vec_words)
```

```python
df2[df2['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = df2[df2['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df2.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```

```python
recommend("Avatar")
```
